```python
import duckdb
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set chart style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

print("Libraries loaded ✅")
```

```python
# Create a persistent database file
con = duckdb.connect('cosmetics.db')

# Load the CSV(s) — this takes ~30-60 seconds
con.execute("""
    CREATE OR REPLACE TABLE events AS
    SELECT * FROM read_csv_auto('data/2019-Oct.csv', header=True)
""")

# If you have both months, use UNION ALL:
# con.execute("""
#     CREATE OR REPLACE TABLE events AS
#     SELECT * FROM read_csv_auto('data/2019-Oct.csv', header=True)
#     UNION ALL
#     SELECT * FROM read_csv_auto('data/2019-Nov.csv', header=True)
# """)

print("Data loaded ✅")
```

```python
# Total rows + date range
result = con.execute("""
    SELECT
        COUNT(*) AS total_events,
        MIN(event_time) AS earliest,
        MAX(event_time) AS latest,
        COUNT(DISTINCT user_id) AS unique_users
    FROM events
""").df()

print(result)
```

```python
# What does an event look like?
con.execute("SELECT * FROM events LIMIT 5").df()
```

```python
# What event types exist?
con.execute("""
    SELECT event_type, COUNT(*) AS count
    FROM events
    GROUP BY event_type
    ORDER BY count DESC
""").df()
```

```python
funnel = con.execute("""
    SELECT
        COUNT(DISTINCT CASE WHEN event_type = 'view'     THEN user_id END) AS viewers,
        COUNT(DISTINCT CASE WHEN event_type = 'cart'     THEN user_id END) AS cart_adders,
        COUNT(DISTINCT CASE WHEN event_type = 'purchase' THEN user_id END) AS buyers
    FROM events
""").df()

# Calculate rates
funnel['view_to_cart_pct'] = round(funnel['cart_adders'] * 100 / funnel['viewers'], 2)
funnel['cart_to_purchase_pct'] = round(funnel['buyers'] * 100 / funnel['cart_adders'], 2)
funnel['overall_conversion_pct'] = round(funnel['buyers'] * 100 / funnel['viewers'], 2)

print(funnel.T)  # transpose for readability
```

```python
funnel = con.execute("""
    SELECT
        COUNT(DISTINCT CASE WHEN event_type = 'view'     THEN user_id END) AS viewers,
        COUNT(DISTINCT CASE WHEN event_type = 'cart'     THEN user_id END) AS cart_adders,
        COUNT(DISTINCT CASE WHEN event_type = 'purchase' THEN user_id END) AS buyers
    FROM events
""").df()

# Calculate rates
funnel['view_to_cart_pct'] = round(funnel['cart_adders'] * 100 / funnel['viewers'], 2)
funnel['cart_to_purchase_pct'] = round(funnel['buyers'] * 100 / funnel['cart_adders'], 2)
funnel['overall_conversion_pct'] = round(funnel['buyers'] * 100 / funnel['viewers'], 2)

print(funnel.T)  # transpose for readability
```

```python
abandonment = con.execute("""
    WITH user_actions AS (
        SELECT
            user_id,
            MAX(CASE WHEN event_type = 'cart'             THEN 1 ELSE 0 END) AS added_cart,
            MAX(CASE WHEN event_type = 'purchase'         THEN 1 ELSE 0 END) AS purchased,
            MAX(CASE WHEN event_type = 'remove_from_cart' THEN 1 ELSE 0 END) AS removed
        FROM events
        GROUP BY user_id
    )
    SELECT
        SUM(added_cart) AS users_added_cart,
        SUM(CASE WHEN added_cart = 1 AND purchased = 0 THEN 1 END) AS abandoned,
        SUM(CASE WHEN added_cart = 1 AND removed = 1   THEN 1 END) AS actively_removed,
        ROUND(SUM(CASE WHEN added_cart = 1 AND purchased = 0 THEN 1 END) * 100.0
              / NULLIF(SUM(added_cart), 0), 2) AS abandonment_rate_pct
    FROM user_actions
""").df()

print(abandonment.T)
```

```python
retention = con.execute("""
    WITH first_seen AS (
        SELECT
            user_id,
            DATE_TRUNC('week', MIN(event_time)) AS cohort_week
        FROM events
        GROUP BY user_id
    ),
    user_weekly_activity AS (
        SELECT DISTINCT
            user_id,
            DATE_TRUNC('week', event_time) AS active_week
        FROM events
    ),
    cohort_activity AS (
        SELECT
            fs.cohort_week,
            DATE_DIFF('week', fs.cohort_week, ua.active_week) AS week_number,
            COUNT(DISTINCT ua.user_id) AS active_users
        FROM first_seen fs
        JOIN user_weekly_activity ua USING (user_id)
        GROUP BY fs.cohort_week, week_number
    ),
    cohort_size AS (
        SELECT cohort_week, COUNT(DISTINCT user_id) AS cohort_users
        FROM first_seen
        GROUP BY cohort_week
    )
    SELECT
        ca.cohort_week,
        cs.cohort_users,
        ca.week_number,
        ca.active_users,
        ROUND(ca.active_users * 100.0 / NULLIF(cs.cohort_users, 0), 2) AS retention_pct
    FROM cohort_activity ca
    JOIN cohort_size cs USING (cohort_week)
    WHERE ca.week_number BETWEEN 0 AND 4
    ORDER BY ca.cohort_week, ca.week_number
""").df()

print(retention.head(20))
```

```python
categories = con.execute("""
    WITH category_metrics AS (
        SELECT
            COALESCE(category_code, 'unknown') AS category,
            COUNT(*) FILTER (WHERE event_type = 'view')     AS views,
            COUNT(*) FILTER (WHERE event_type = 'purchase') AS purchases,
            ROUND(AVG(price)::numeric, 2)                   AS avg_price
        FROM events
        WHERE category_code IS NOT NULL
        GROUP BY category
        HAVING COUNT(*) FILTER (WHERE event_type = 'view') > 5000
    )
    SELECT
        category,
        views,
        purchases,
        avg_price,
        ROUND(views * 1.0 / NULLIF(purchases, 0), 2) AS views_per_purchase,
        ROUND(purchases * 100.0 / NULLIF(views, 0), 2) AS conversion_rate_pct
    FROM category_metrics
    ORDER BY views_per_purchase DESC
    LIMIT 10
""").df()

print(categories)
```

```python
import matplotlib.pyplot as plt

stages = ['Viewed', 'Added to Cart', 'Purchased']
values = [funnel['viewers'][0], funnel['cart_adders'][0], funnel['buyers'][0]]
colors = ['#3498db', '#f39c12', '#2ecc71']

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(stages, values, color=colors, edgecolor='black', linewidth=1.2)

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{int(height):,}',
                xy=(bar.get_x() + bar.get_width()/2, height),
                xytext=(0, 5), textcoords='offset points',
                ha='center', fontsize=12, fontweight='bold')

ax.set_ylabel('Unique Users', fontsize=12)
ax.set_title('Conversion Funnel — Cosmetics Store, October 2019',
             fontsize=14, fontweight='bold')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('charts/01_funnel.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Pivot to make it chartable
pivot = retention.pivot(index='cohort_week', columns='week_number', values='retention_pct')

fig, ax = plt.subplots(figsize=(12, 6))

for cohort in pivot.index[:5]:  # show first 5 cohorts
    cohort_label = pd.to_datetime(cohort).strftime('%b %d')
    ax.plot(pivot.columns, pivot.loc[cohort], marker='o',
            linewidth=2, label=f'Week of {cohort_label}')

ax.set_xlabel('Weeks Since First Activity', fontsize=12)
ax.set_ylabel('Retention %', fontsize=12)
ax.set_title('Weekly Cohort Retention', fontsize=14, fontweight='bold')
ax.legend(loc='upper right', framealpha=0.9)
ax.grid(True, alpha=0.3)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('charts/02_retention.png', dpi=150, bbox_inches='tight')
plt.show()
```
